```python
#Bar Graph
plt.figure(figsize=(10,10)) #size
age=[34,36,41,43,28]
name=['vk','sky','msd','st','hp']
plt.bar(name,age)
```

```python
#changing size
plt.figure(figsize=(9,4)) #size
age=[34,36,41,43,28]
name=['vk','sky','msd','st','hp']
plt.bar(name,age)
```

```python
#stacked bar chart
name=['shiva','vandu','ram','ravi']
ps=[29,48,62,36]
ns=[14,45,32,22]
plt.bar(name,ps,label='physics',color='b')
plt.bar(name,ns,label='biology',color='g')
```

```python
#labeling
name=['shiva','vandu','ram','ravi']
ps=[29,48,62,36]
ns=[14,45,32,22]
plt.bar(name,ps,label='physics',width=0.75,color='k')
plt.bar(name,ns,label='biology',width=0.75,color='g')
plt.xlabel('name')
plt.ylabel('marks')
plt.legend()
```

```python
#pie chart
age=[34,36,41,43,28]
name=['vk','sky','msd','st','hp']
clr=['g','b','r','k','m']
plt.pie(age,labels=name,colors=clr,startangle=0)
```

```python
#bar graph
x=[1,2,3,4,5,6,7,8]
y=[5,9,6,3,7,8,4,1]
plt.bar(x,y)
```

```python
#line graph
plt.plot(x,y)
```

```python
#scatter graph
plt.scatter(x,y)
```

```python
#subplot(multigraphs in one figure)
fig,ax=plt.subplots(2,2) #since 2 by 2 subplot
ax[0,0].plot(x,y)
ax[0,1].pie(data)
ax[1,0].bar(x,y)
ax[1,1].scatter(x,y)
```

```python
#size adjusting
rcParams['figure.figsize']=8,4
fig,ax=plt.subplots(2,2)
ax[0,0].plot(x,y)
ax[0,1].pie(data)
ax[1,0].bar(x,y)
ax[1,1].scatter(x,y)
```
